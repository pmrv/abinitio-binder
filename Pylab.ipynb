```python
import numpy as np
import matplotlib.pyplot as plt
```

[Numpy](https://numpy.org/) is a library to simplify and speed up array manipulations and operations in Python.  Below is a short intro, but also have a look at the [official intro](https://numpy.org/devdocs/user/quickstart.html).

There a few ways to create a new numpy array and the three most important are shown below. `array()` takes an existing python list (can also be a nested list) and returns an array with the same content; `arange()` works like the standard python function [range()](https://docs.python.org/3/library/stdtypes.html?highlight=range#range), but returns a numpy array, while the `linspace()` gives an array of evenly spaced samples in the specified range.  For more details also check [Array Creation Routines](https://numpy.org/devdocs/reference/routines.array-creation.html) in the official manual.

```python
a = np.array([0, 1, 2, 3, 4])
b = np.arange(5)
c = np.linspace(0, 1, 50)
print(a)
print(b)
print(c)
```

The main advantage for us is how much more convenient numpy arrays are than standard lists, e.g. if we wanted to multiply the values in two lists

```python
a = [10, 15, 31, 27]
b = list(range(4))
c = [0, 0, 0, 0]

for i in range(len(a)):
    c[i] = a[i] * b[i]
    
print(a, b, c)
```

```python
na = np.array([10, 15, 31, 27])
nb = np.arange(4)
nc = na * nb

print(na, nb, nc)
```

Numpy arrays allows all kinds of arithmetic to be applied element-wise to the array like this

```python
print(na + nb)
print( (na - nb) / na )
print(nc ** nb)
```

In the same way mathematical functions that are applied to arrays are applied element-wise and numpy defines [a lot](https://numpy.org/devdocs/reference/routines.math.html) of them.

```python
x = np.linspace(0, 2 * np.pi, 10)
print(x, np.sin(x))
print(x, np.exp(x))
```

Instead of just producing numbers it'd also be nice to plot them in a convenient way.  For this reason [matplotlib](matplotlib.org) exists (it also comes with its own [documentation](https://matplotlib.org/tutorials/index.html)).  The central function we'll need is called `plot()` and takes the x values and the y values as two separate arguments, like this

```python
x = np.linspace(0, 2 * np.pi, 100)
plt.plot(x, np.sin(x))
```

To plot multiple functions at once, either put them on after another in the same call to `plot()` or call `plot()` multiple times in the same cell.

```python
plt.plot(x, np.sin(x), x, np.cos(x))
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

You can also change the formatting of the plot by add a third argument.  This argument is a special string that tells matplotlib what color and marker to use for the plot, e.g.

```python
x = np.linspace(0, 2 * np.pi, 10)
plt.plot(x, np.sin(x), '-') # a simple line, the default
plt.plot(x, np.sin(x), 'o') # circle markers
```

In the same string you can also add letters to specify colors or combine multiple markers.

```python
plt.plot(x, np.sin(x), 'bo-') # blue circles and line
```
